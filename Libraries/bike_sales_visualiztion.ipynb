## Feature Engineering

```python
Sales['Quarter'] = pd.to_datetime(Sales['Date']).dt.quarter
Sales['Quarter']
```

```python
# check the day of the week for each date and calculate if it's a weekend or not
Sales['Day_Name'] = pd.to_datetime(Sales['Date']).dt.day_name()
Sales['Is_Weekend'] = Sales['Day_Name'].isin(['Saturday','Sunday']).astype(int)
```

```python
# create a function to determine the season based on the month
def get_season(month):
    if month in ['December','January','February']:
        return 'Winter'
    elif month in ['March','April','May']:
        return 'Spring'
    elif month in ['June','July','August']:
        return 'Summer'
    else:
        return 'Fall'

Sales['Season'] = Sales['Month'].apply(get_season)
```

```python
# Calculate profit margin
Sales['Profit_Margin'] = Sales['Profit'] / Sales['Revenue']
```

```python
# Calculate revenue per unit
Sales['Revenue_per_Unit'] = Sales['Revenue'] / Sales['Order_Quantity']
```

```python
Sales['Cost_per_Unit'] = Sales['Cost'] / Sales['Order_Quantity']
```

```python
Sales['Age_Bucket'] = pd.cut(Sales['Customer_Age'],
                          bins=[0,25,35,50,70],
                          labels=['Youth','Young Adult','Mid Age','Senior'])
```

```python
product_avg_rev = Sales.groupby('Product')['Revenue'].mean()
Sales['Product_Avg_Revenue'] = Sales['Product'].map(product_avg_rev)
```

```python
# create combined ageâ€‘group/product category column on Sales dataframe
Sales['Age_Category'] = Sales['Age_Group'] + "_" + Sales['Product_Category']
```

```python
# Calculate cost efficiency as profit divided by cost
Sales['Cost_Efficiency'] = Sales['Profit'] / Sales['Cost']
```

```python
#Revenue Growth Over Years
year_rev = Sales.groupby('Year')['Revenue'].sum()
```

```python
# Create a new column 'High_Value_Order' which is 1 if Revenue is above the median, else 0
Sales['High_Value_Order'] = (Sales['Revenue'] > Sales['Revenue'].median()).astype(int)
```

## Visualizations 

```python
# Visualize the distribution of orders by country
plt.figure(figsize=(10,5))
sns.countplot(data=Sales, x="Country", order=Sales["Country"].value_counts().index)
plt.title("Number of Orders by Country")
plt.xticks(rotation=45)
plt.show()
```

Insight:

- United States and Australia generate the highest revenue among all countries.

- This suggests strong market demand in these regions.

- The company may consider increasing marketing investment in top-performing countries.

- Low-performing countries might require promotional strategies or pricing review.

```python
#Revenue by Product Category
#Which category generates the highest revenue?
plt.figure(figsize=(8,5))
sns.barplot(data=Sales, x="Product_Category", y="Revenue", estimator=sum)
plt.title("Total Revenue by Product Category")
plt.show()
```

Insight:

- The category generating the highest revenue is likely Bikes 

- However, high revenue does not always mean high profitability.

- Clothing may have lower revenue but potentially higher profit margins.

- Revenue alone does not reflect business efficiency; profit margin must also be considered.

Insight:

```python
#Best and worst months for sales
monthly_revenue = Sales.groupby("Month")["Revenue"].sum().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_revenue, x="Month", y="Revenue", marker="o")
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
#Profit vs Cost (Relationship)
#Higher cost â†’ higher profit?
plt.figure(figsize=(7,5))
sns.scatterplot(data=Sales, x="Cost", y="Profit", hue="Product_Category")
plt.title("Profit vs Cost")
plt.show()
```
